# 📘 Projet 3 : Éducation
## 🎯 Analyse des performances académiques des étudiants

Ce projet a pour objectif d’identifier les facteurs
qui influencent la réussite académique des étudiants
à partir d’un **Student Performance Dataset**.

---

## 📂 Dataset utilisé
Le dataset contient plusieurs variables explicatives :

- **Gender**
- **ParentalEducation**
- **StudyTimeWeekly**
- **Absences**
- **Tutoring**
- **ParentalSupport**
- **GPA**
- **GradeClass**

---

## 🎯 Problématique
Les performances des étudiants diffèrent-elles
significativement selon la méthode d’apprentissage utilisée ?

### Questions secondaires
- Le genre influence-t-il le GPA ?
- Le niveau d’éducation des parents a-t-il un impact ?
- Le temps d’étude améliore-t-il la réussite ?
- Les absences réduisent-elles les performances ?

---

## 🧪 Méthodes utilisées
Le projet utilise :

- **Python**
- **Pandas**
- **Matplotlib**
- **Scipy**
- **ANOVA**
- **EDA (Exploratory Data Analysis)**

---

## 📊 Analyses réalisées
- Impact du **Tutoring** sur le GPA
- Comparaison selon le **genre**
- Effet du **niveau parental**
- Relation **Study Time vs GPA**
- Corrélations entre variables
- Test **ANOVA**

---

## ✅ Résultats principaux
Les résultats montrent que :

- le **tutoring améliore significativement le GPA**
- plus le temps d’étude augmente, plus le GPA augmente
- les absences réduisent les performances
- le soutien parental améliore les résultats

👉 La méthode la plus efficace combine :

> **Tutoring + temps d’étude élevé + faible nombre d’absences**
